## Extração dos dados do Enade

```python
# importar bibliotecas necessarias 
import pandas as pd
import numpy as np
import zipfile
import requests
from io import BytesIO
import os
```

```python
# Criar um diretorio para armazenar o conteudo do enades
os.makedirs('./enade2019', exist_ok=True)
```

```python
# Define a url 
url = "https://download.inep.gov.br/microdados/Enade_Microdados/microdados_enade_2019.zip"

#Faz download do conteudo 
filebytes = BytesIO(
    requests.get(url).content
)

# Extrair o conteudo do zipefile
myzip = zipfile.ZipFile(filebytes)
myzip.extractall("./enade2019")
```
